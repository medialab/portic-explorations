```python
# bootstrapping du client
from lib.client import Api
import json

client = Api()
```

```python
def query_larochellepartners(start="1787", end="1789", source="National toutes directions partenaires manquants", classification='partner_simplification'):
    
    flux = client.toflit.get_locations(classification, {
        "partnerClassification":None,
        "kind":"total",
        "sourceType":source,
        "dateMin":start,
        "dateMax":end
    })

    def filter_to_larochelle (f):
        if (f['direction'] == 'La Rochelle'):
            return True
        else:
            return False
    # print("")
    print ("Recherche de partenaires de la direction de la Rochelle de " + start + " à " + end + " selon la source " + source + "(classification : " + classification + ")")
    # print("")
    if (flux):
        flux = filter(filter_to_larochelle, flux)
        # for f in list(flux):
        #     print( f['partner'] + '(' + str(f['count']) + ' flux)')
        return list(flux)
    else :
        # print("Pas de résultats")
        return None
        
# query_larochellepartners(start="1787", end="1789", source="National toutes directions partenaires manquants", classification='partner_simplification')

sources = client.toflit.get_sources_types()
partners_classifications = ["partner_source","partner_orthographic","partner_simplification","partner_africa","partner_grouping","partner_obrien","partner_sourcename","partner_wars"]
years = ["1787", "1788", "1789"]

results = list()
for year in years:
    for source in sources:
        for partner_classification in partners_classifications:
            matching = query_larochellepartners(start=year, end=year, source=source, classification=partner_classification)
            if (matching is not None):
                print("===")
                print("match")
                print("===")
                results.append({
                    "year":year,
                    "classification":partner_classification,
                    "source":source,
                    "flux":matching
                })
```

```python
# formattage des résultats
results_filtered = filter(lambda r : len(r["flux"]) > 0, results)

print("| année | source | classification | nombre de flux | détail des flux |")
print("|---|---|---|---|---|")
for item in list(results_filtered):
    print('| ' + str(item["year"]) + ' | ' + item["source"] + ' | ' + item["classification"] + ' | ' + str(len(item["flux"])) + '|' + ', '.join(list(map(lambda i : i["partner"] + "(" + str(i["count"]) + ")", item["flux"]))) + ' |')
    # print('| ' + str(item["year"]) + ' | ' + item["source"] + " | " + item["classification"] + len(item["flux"]) ± " | |")    
```
